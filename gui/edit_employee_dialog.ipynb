```python
# gui/edit_employee_dialog.py
import tkinter as tk
from tkinter import ttk, messagebox
from models.validation import validate_email, validate_dob

class EditEmployeeDialog(tk.Toplevel):
    def __init__(self, master, db, emp):
        super().__init__(master)
        self.db = db
        self.emp = emp
        self.title("Edit Employee")
        self.geometry("400x400")
        self.create_widgets()

    def create_widgets(self):
        labels = ["First Name", "Last Name", "Email", "Pay Type", "DOB (YYYY-MM-DD)"]
        self.entries = {}
        for label in labels:
            ttk.Label(self, text=label).pack(pady=5)
        self.entries["First Name"] = ttk.Entry(self)
        self.entries["First Name"].pack(pady=5)
        self.entries["First Name"].insert(0, self.emp.first_name)
        self.entries["Last Name"] = ttk.Entry(self)
        self.entries["Last Name"].pack(pady=5)
        self.entries["Last Name"].insert(0, self.emp.last_name)
        self.entries["Email"] = ttk.Entry(self)
        self.entries["Email"].pack(pady=5)
        self.entries["Email"].insert(0, self.emp.email)
        self.entries["Pay Type"] = ttk.Entry(self)
        self.entries["Pay Type"].pack(pady=5)
        self.entries["Pay Type"].insert(0, self.emp.pay_type)
        self.entries["DOB (YYYY-MM-DD)"] = ttk.Entry(self)
        self.entries["DOB (YYYY-MM-DD)"].pack(pady=5)
        self.entries["DOB (YYYY-MM-DD)"].insert(0, self.emp.dob)
        ttk.Button(self, text="Save", command=self.save_changes).pack(pady=10)

    def save_changes(self):
        first = self.entries["First Name"].get()
        last = self.entries["Last Name"].get()
        email = self.entries["Email"].get()
        pay_type = self.entries["Pay Type"].get()
        dob = self.entries["DOB (YYYY-MM-DD)"].get()

        if not validate_email(email):
            messagebox.showerror("Error", "Invalid email format")
            return
        if not validate_dob(dob):
            messagebox.showerror("Error", "Employee must be at least 18 years old")
            return

        self.emp.first_name = first
        self.emp.last_name = last
        self.emp.email = email
        self.emp.pay_type = pay_type
        self.emp.dob = dob

        self.db.save_employees()
        messagebox.showinfo("Success", "Employee updated")
        self.destroy()
```
