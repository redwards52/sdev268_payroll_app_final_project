```python
# gui/login_window.py
import tkinter as tk
from tkinter import ttk, messagebox
from models.database import Database
from ui.main_gui import MainGUI

class LoginWindow(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title("Payroll Login")
        self.geometry("300x200")
        self.db = Database()
        self.create_widgets()

    def create_widgets(self):
        ttk.Label(self, text="User ID").pack(pady=5)
        self.user_entry = ttk.Entry(self)
        self.user_entry.pack(pady=5)

        ttk.Label(self, text="Password").pack(pady=5)
        self.pass_entry = ttk.Entry(self, show="*")
        self.pass_entry.pack(pady=5)

        ttk.Button(self, text="Login", command=self.login).pack(pady=10)

    def login(self):
        uid = self.user_entry.get()
        pwd = self.pass_entry.get()
        if uid == "HR0001" and pwd == "adminpass":  # Example; later hash it
            self.destroy()
            MainGUI(self.db)
        else:
            messagebox.showerror("Login Failed", "Invalid credentials")
```
