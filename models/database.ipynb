```python
# models/database.py
from models.employee import Employee

class Database:
    def __init__(self):
        self.employees = []

    def add_employee(self, emp: Employee):
        if any(e.employee_id == emp.employee_id for e in self.employees):
            raise ValueError(f"Employee ID {emp.employee_id} already exists.")
        self.employees.append(emp)

    def list_employees(self):
        return self.employees

    def delete_employee(self, emp_id: str):
        self.employees = [e for e in self.employees if e.employee_id != emp_id]

    def search_employee(self, term: str):
        for e in self.employees:
            if term.lower() in (e.employee_id.lower(), e.email.lower()):
                return e
        return None
```
